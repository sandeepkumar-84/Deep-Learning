#Churn prediction using multi layer perceptron or ANN
# since it is a binary classification problem, the activation in output layer is sigmoid which gives probabilities (or it could be a step function that gives 0 or 1) and loss function is binary_crossentropy

```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rjmanoj/credit-card-customer-churn-prediction")

print("Path to dataset files:", path)
```

```python
import os

# Construct the full path to the CSV file
csv_file_path = os.path.join(path, 'Churn_Modelling.csv')

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
display(df.head())
```

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], inplace = True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True )
```

```python
df.head()
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(11,activation = 'sigmoid', input_dim = 11))
model.add(Dense(3,activation = 'sigmoid'))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#model.fit(X_train_scaled,y_train, epochs = 100,validation_split = 0.2)
model.fit(X_train_scaled,y_train, batch_size=50, epochs = 100,validation_split = 0.2, verbose = 1)
```

```python
model.layers[1].get_weights()
```

```python
y_log = model.predict(X_test_scaled)
```

```python
import numpy as np
y_pred = np.where(y_log > 0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```
